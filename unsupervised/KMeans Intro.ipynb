```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
%matplotlib inline

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("delivery_fleet.csv", delimiter="\t")
df.head()
```

```python
f1 = df["Distance_Feature"]
f2 = df["Speeding_Feature"]
```

```python
plt.scatter(f1,f2, color="k")
plt.xlabel("Distance_Feature")
plt.ylabel("Speeding_Feature")
```

```python
X = df[["Distance_Feature", "Speeding_Feature"]]
```

```python
kmeans = KMeans(n_clusters=2).fit(X)
kmeans.cluster_centers_
```

```python
X.head()
```

```python
for i in xrange(0, X.shape[0]):
    if kmeans.labels_[i] == 0:
        c1 = plt.scatter(X.iloc[i,0],X.iloc[i,1],c='r', marker='+')
    elif kmeans.labels_[i] == 1:
        c2 = plt.scatter(X.iloc[i,0], X.iloc[i,1], c='b', marker='*')
plt.title("Kmean. n_clusters=2")
```

```python
kmeans = KMeans(n_clusters=4).fit(X)
```

```python
for i in xrange(0, X.shape[0]):
    if kmeans.labels_[i] == 0:
        c1 = plt.scatter(X.iloc[i,0],X.iloc[i,1],c='r', marker='+')
    elif kmeans.labels_[i] == 1:
        c2 = plt.scatter(X.iloc[i,0], X.iloc[i,1], c='b', marker='*')
    elif kmeans.labels_[i] == 2:
        c3 = plt.scatter(X.iloc[i,0], X.iloc[i,1], c='k', marker='o')
    elif kmeans.labels_[i] == 3:
        c4 = plt.scatter(X.iloc[i,0], X.iloc[i,1], c='g', marker='^')
plt.title("Kmean. n_clusters=4")
```
